```python
import pandas as pd
metrics = pd.read_csv('metrics.csv')
metrics.sort_values('testr2', ascending = False)
```

```python
metrics.groupby(['HL1', 'HL2'])['testr2'].describe()
```

```python
theekthaakmetrics = metrics[metrics["testr2"] > 0].copy()
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("darkgrid")
sns.FacetGrid(theekthaakmetrics, hue = None, height = 10)\
    .map(plt.scatter, "HL1", "testr2")\
    .add_legend()
plt.show()
sns.set_style("darkgrid")
sns.FacetGrid(theekthaakmetrics, hue = None, height = 10)\
    .map(plt.scatter, "HL2", "testr2")\
    .add_legend()
plt.show()
sns.set_style("darkgrid")
sns.FacetGrid(theekthaakmetrics, hue = None, height = 10)\
    .map(plt.scatter, "learning", "testr2")\
    .add_legend()
plt.show()
sns.set_style("darkgrid")
sns.FacetGrid(theekthaakmetrics, hue = None, height = 10)\
    .map(plt.scatter, "max_iter", "testr2")\
    .add_legend()
plt.show()
sns.set_style("darkgrid")
sns.FacetGrid(theekthaakmetrics, hue = None, height = 10)\
    .map(plt.scatter, "n_iter_no_change", "testr2")\
    .add_legend()
plt.show()
```
